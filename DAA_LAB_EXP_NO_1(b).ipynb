```python
import random

def interpolation_search(arr, target):
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high and target >= arr[low] and target <= arr[high]:
        comparisons += 1

        if arr[high] == arr[low]:
            break

        pos = low + int(
            ((target - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        if pos < low or pos > high:
            break

        if abs(arr[pos] - target) < 1e-9:
            return pos, comparisons

        elif arr[pos] < target:
            low = pos + 1

        else:
            high = pos - 1

    return -1, comparisons


# Dataset sizes
sizes = [10000, 50000, 100000]

print("Interpolation Search Analysis")
print("-" * 40)

for size in sizes:

    # Generate uniformly distributed floating-point numbers
    arr = sorted(random.uniform(0.0, 1000.0) for _ in range(size))

    # Select a target from the array
    target = arr[size // 2]

    index, comparisons = interpolation_search(arr, target)

    print(f"\nDataset Size: {size}")
    print(f"Target Value: {target:.4f}")
    print(f"Index Found: {index}")
    print(f"Comparisons Required: {comparisons}")
```
